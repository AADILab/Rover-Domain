```python
fig, ax = plt.subplots()
for key in ("G/three_poi/two_agent", "G/three_poi/two_agent", "hierarchy/three_poi/two_agent", "hierarchy/three_poi/three_agent"):
    df = pd.read_hdf("hierarchy-multi-reward_best.h5", key)
    df.mean()[::25].plot(ax=ax, yerr=df.std()[::25]/np.sqrt(10), label=key)
    print(key, df.mean().mean(), df.mean().std())
ax.set_title("Three POI case")
ax.legend(loc="lower left")
ax.set_ylim(-0.2, 3.2)
ax.set_ylabel("Reward (Progression through graph)")
ax.set_xlabel("Generation")
```
